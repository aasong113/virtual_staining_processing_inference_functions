```python
import os

def rename_images_with_prefix(folder_path, prefix="kidney_normal_1_"):
    for filename in os.listdir(folder_path):
        if filename.lower().endswith(('.tif', '.tiff', '.png', '.jpg', '.jpeg')):
            old_path = os.path.join(folder_path, filename)
            new_filename = prefix + filename
            new_path = os.path.join(folder_path, new_filename)
            os.rename(old_path, new_path)
            print(f"Renamed: {filename} -> {new_filename}")

```

```python
import os
import shutil

def move_images(folder1, folder2, extensions={".jpg", ".jpeg", ".png", ".tif", ".tiff", ".bmp"}):
    if not os.path.exists(folder2):
        os.makedirs(folder2)
    
    for filename in os.listdir(folder1):
        if any(filename.lower().endswith(ext) for ext in extensions):
            src_path = os.path.join(folder1, filename)
            dst_path = os.path.join(folder2, filename)
            try:
                shutil.move(src_path, dst_path)
                print(f"Moved: {filename}")
            except Exception as e:
                print(f"Error moving {filename}: {e}")
```
